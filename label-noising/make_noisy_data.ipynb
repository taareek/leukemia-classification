```python
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import matthews_corrcoef
from imblearn.metrics import specificity_score
```

```python
def eval_results(model, tain_x, train_y, test_x, test_y):
  model.fit(train_x, train_y)
  prediction = model.predict(test_x)
  acc = metrics.accuracy_score(prediction, test_y)
  print(f"Accuracy: {acc}")
  roc_auc = roc_auc_score(test_y, prediction)
  print("ROC-AUC score is: ", round(roc_auc, 3))
  mcc = matthews_corrcoef(test_y, prediction)
  print("The Matthews correlation coefficinet (MCC) is: ", round(mcc, 3))
  sp = specificity_score(test_y, prediction, average='weighted')
  print(f"Weighted specificity: {sp:.3f}")
  print(classification_report(test_y, prediction))
```

```python
# adaboost classifier
from sklearn.ensemble import AdaBoostClassifier

adaboost_clf = AdaBoostClassifier(n_estimators=100, learning_rate=1.0, algorithm='SAMME.R', random_state=42, base_estimator='deprecated')
eval_results(adaboost_clf, train_x, train_y, test_x, test_y)
```

```python
# functio to make noisy labels
def make_noisy_label(true_labels, dummy_labels, n_samples):
  k = 0
  for i in range(len(true_labels)):
    if k>= n_samples:
      dummy_labels[i] = true_labels[i]
    else:
      if(true_labels[i]==0):
        dummy_labels[i] = 1
      elif(true_labels[i]==1):
        dummy_labels[i] = 0
    k+= 1
  return dummy_labels
```

```python
noise_train = make_noisy_label(train_f, train_dummy, n_5_percentage)
print(f"shape of lable_noise_5: {noise_train.shape} and unique values are: {np.unique(noise_train)}")
```
